**1_ Importar el algoritmo y librerias necesarias**

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

**2_ Selección de Datos. Obtener los datos usando la librería Pandas**

Utilizo la libreria Pandas para importar el archivo de origen
Revisar el archivo de origen para ver el formato. En este caso, delimitado por comas y los decimales con punto. 

El método read_csv de Pandas devuelve un DataFrame.

**Datos**

```python
datos= pd.read_csv('data/AprendizajeCredito.csv',delimiter=',',decimal='.')
```

```python
y=datos.iloc[:,6:7]
```

Seleccionamos un 35% de los valores para testear y un 65% de ellos para el entrenamiento.

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.35,random_state=0)
```

```python
type(x_train)
```

```python
x_train.head(57)
```

```python
type(y_train)
```

```python
y_train.head(86)
```

Creamos instancia del clasificador

```python
instancia_arbol = tree.DecisionTreeClassifier()
```

Entrenamos con FIT

```python
instancia_arbol.fit(x_train,y_train)
```

Predicciones:

```python
prediccion_test = instancia_arbol.predict(x_test)
```

```python
type(prediccion_test)
```

```python
prediccion_test_df = pd.DataFrame(prediccion_test)
```

```python
prediccion_test_df.head(179)
```

Clasificación de la instancia.

```python
print("Es buen pagador: " + instancia_arbol.predict([[1,3,4,2,1]]))
```

```python
print("Es buen pagador: " + instancia_arbol.predict([[3,2,2,1,3]]))
```

Procedemos a calcular el porcentaje de clasificación global

```python
porcentaje_predict_global=instancia_arbol.score(x_test,y_test)
print("El porcentaje es:")
print(porcentaje_predict_global)
```
